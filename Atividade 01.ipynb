10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
#Instalar a biblioteca para trabalhar com banco de dados, se ja estiver essa biblioteca instalada, nao 
#executar esta etapa
!pip install sqlalchemy==1.4.46
```

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect

#colocar na pasta notebooks_aulas, nao consegui executar com o arquivo na pasta notebooks_atividades
engine = create_engine('sqlite:///dados/aula3/sqlite-sakila.db')
insp = inspect(engine)
```

```python
insp.get_table_names()
```

```python
#Select entre as tabelas city e country para listar as cidades dos paises que comecam com B no nome
import pandas as pd

sql_query ='''
select
    ci.city,
    co.country
from 
    city ci
inner join country co
    on ci.country_id = co.country_id
where co.country like "B%"
 '''
country_city_df = pd.read_sql_query(sql_query, engine)
```
